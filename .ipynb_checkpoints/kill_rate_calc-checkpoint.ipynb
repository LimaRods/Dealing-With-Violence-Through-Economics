```python
import pandas as pd
```

```python
def get_year(date):
    return int(str(date)[-4:])
```

```python
def concat_codes(row):
    final = row.COD_UF + row.COD_MUN
    return final[:-1]
```

# 1. Tratar Dados de População

```python
df_pop = pd.read_excel("data/POP_MUN.xls", dtype = str)
df_pop['COD_MUN_completo'] = df_pop.apply(concat_codes, axis=1)
df_pop.rename({"COD_MUN_completo":"COD_MUNICIPIO"}, inplace=True, axis=1)
df_pop = df_pop[['COD_MUNICIPIO', 'POP_ESTIM']]
```

# 2. Tratar dados de óbitos

```python
df_mortes = pd.read_csv("datas/SIM_homicidios.csv")
df_mortes['year_obito'] = df_mortes['DTOBITO'].apply(get_year)
```

# 2.1 Óbitos Totais e associar à tabela final

```python
mortes_per_mun = pd.DataFrame(df_mortes.groupby(by=['CODMUNOCOR', 'year_obito'])['CONTADOR'].count())
mortes_per_mun.rename({"CONTADOR":"num_mortes"}, axis=1, inplace=True)
mortes_per_mun.reset_index(inplace=True)
mortes_per_mun.rename({"CODMUNOCOR":"COD_MUNICIPIO"}, axis=1, inplace=True)
mortes_per_mun['COD_MUNICIPIO'] = mortes_per_mun['COD_MUNICIPIO'].apply(str)
```

```python
df_mortes_pop = mortes_per_mun.merge(df_pop, on="COD_MUNICIPIO", how='left')
df_mortes_pop['num_mortes'] = df_mortes_pop['num_mortes'].astype(float)
df_mortes_pop['POP_ESTIM'] = df_mortes_pop['POP_ESTIM'].astype(float)
```

```python
df_mortes_pop
```

# 2.2 Óbitos por Homicídio e Associar População

```python
df_hom = df_mortes[df_mortes['CIRCOBITO'] == 3]
```

```python
hom_per_mun = pd.DataFrame(df_hom.groupby(by=['CODMUNOCOR', 'year_obito'])['CONTADOR'].count())
hom_per_mun.rename({"CONTADOR":"num_hom"}, axis=1,inplace=True)
hom_per_mun.reset_index(inplace=True)
hom_per_mun.rename({"CODMUNOCOR":"COD_MUNICIPIO"}, axis=1,inplace=True)
hom_per_mun['COD_MUNICIPIO'] = hom_per_mun['COD_MUNICIPIO'].apply(str)
```

```python
hom_per_mun
```

# 3. Tabela Final dos Targets

```python
df_hom_mortes_pop = df_mortes_pop.merge(hom_per_mun, on=['COD_MUNICIPIO', 'year_obito'], how='left')
df_hom_mortes_pop['num_hom'] = df_hom_mortes_pop['num_hom'].fillna(0)
df_hom_mortes_pop['TX_hom_por_pop [%]'] = (df_hom_mortes_pop['num_hom']/df_hom_mortes_pop['POP_ESTIM'])*100
df_hom_mortes_pop['TX_hom_por_obito [%]'] = (df_hom_mortes_pop['num_hom']/df_hom_mortes_pop['num_mortes'])*100
df_hom_mortes_pop = df_hom_mortes_pop.dropna()
```

```python
df_hom_mortes_pop
```

```python
df_hom_mortes_pop.to_csv("data/TARGETS.csv", index=False)
```
